# Define NumPy array using List and also using Tuple

```python
an_array = np.array([1, 3, 5, 7, 9])

print(an_array)
#Data type of array
print(an_array.dtype)
```

```python
an_array = np.array((1, 3, 5, 7, 9))

print(an_array)
print(type(an_array))
```

# One Dimension Array

```python
one_d_array = np.array([1,3,5,7,9])

print(f"Number of Dimensions:: ${one_d_array.ndim}")
print(one_d_array[1])
```

# Two Dimension Array

```python
two_d_array = np.array([[1,3,5],[7,9,10]])

print(f"Number of Dimensions:: ${two_d_array.ndim}")
print(two_d_array[1, 2])
```

# Three Dimension Array

```python
three_d_array = np.array([[[1,3,5],[7,9,10]],[[2,5,7],[4,5,9]]] )

print(f"Number of Dimensions:: ${three_d_array.ndim}")
print(three_d_array[0, 1, 2])
print(f"Shape: ${three_d_array.shape}")
```

# Size and No of bytes in array

```python
print(f"Size of array:: ${three_d_array.size}")
print(f"Size od array in Bytes:: ${three_d_array.nbytes}")
```
