# Universal Method: np.arange



*   Create an array of 100 elemets
*   Create an array start with 20 and ends at 100
*   Create an array start with 20 and ends at 100 and perform jumps of 20.
*   arange method doesn't create a list. It iterates and gives you numbers. These kind of objects are iterators. They iterate but never create.
*   But you can create a list by calling list method




```python
arange_array = np.arange(100)
print(arange_array)

arange_array = np.arange(20,100)
print(arange_array)

arange_array = np.arange(20,100,20)
print(arange_array)

arange_list = list(np.arange(20,100,20))
print(arange_list)
```

# NumPy.random.rand

# How You read multi-dimension array


*   Let's assume the 4-dimensions array dimensions are (2,3,4,5)
*   There are two arrays that are three dimensional
*   Each three dimensional array has 3 two-dimensional array
*   Each two dimensional array has 4 one-dimensional array
*   Each one dimensional array has 5 elements



```python
rand_array = np.random.rand(10)
print(rand_array)
print("------------------------------")
rand_two_d_array = np.random.rand(2,3)
print(rand_two_d_array)
print("------------------------------")
rand_three_d_array = np.random.rand(2,3,4)
print(rand_three_d_array)
print(rand_three_d_array.ndim)
print("------------------------------")
rand_four_d_array = np.random.rand(2,3,4,5)
print(rand_four_d_array)
print(rand_four_d_array.ndim)
```

# NumPy.reshape


*   This changes the shape of the array
*   An array of shape 1-dimension can be changed to 2 dimension array. Provided the count of the elements supports that.
*   If you create a one-dimensional array of 100 elements. Reshape it to two-dimensional array which would be matrix. Give arguments (4, 25).
*   Now this is two dimensional array and it's matrix
*   4 means rows in the matrix and 25 means columns in the matrix



```python
create_matrix = np.arange(100).reshape(4, 25)

print(create_matrix.shape)
print("-------------------------")
print("4 rows and 25 columns of the matrix")
print("-------------------------")
print(create_matrix)
```

# NumPy.reshape


*   This changes the shape of the array

*   If elements are total hundered, then multiple of all dimensions has to be hundred.

```python
create_three_d_array = np.arange(100).reshape(4, 5, 5)
print(create_three_d_array)
```

# Create a matrix

```python
a_matrix = np.round(10 * np.random.rand(5, 4))
print(a_matrix)
print("-------------------------")
print(a_matrix.shape)
print("-------------------------")
print(a_matrix[1,2])
print("------------Read Complete row 1-------------")
print(a_matrix[1,:])
print("------------Read Complete column 1-------------")
print(a_matrix[:,1])
```

# Matrix Transpose

```python
a_transpose = a_matrix.T
print(a_transpose)
print("-------------------------")
print(a_transpose.shape)
print("-------------------------")
print(a_transpose[1,2])
print("------------Read Complete row 1-------------")
print(a_transpose[1,:])
print("------------Read Complete column 1-------------")
print(a_transpose[:,1])
```

# In matrix, sort all the columns

```python
a_transpose.sort(axis = 0)
print(a_transpose)
print("-------------------------")
print(a_transpose.shape)
print("-------------------------")
print(a_transpose[1,2])
print("------------Read Complete row 1-------------")
print(a_transpose[1,:])
print("------------Read Complete column 1-------------")
print(a_transpose[:,1])
```

# In matrix, sort all the rows

```python
a_transpose.sort(axis = 1)
print(a_transpose)
print("-------------------------")
print(a_transpose.shape)
print("-------------------------")
print(a_transpose[1,2])
print("------------Read Complete row 1-------------")
print(a_transpose[1,:])
print("------------Read Complete column 1-------------")
print(a_transpose[:,1])
```

# Array Masking OR Fancy Indexing


*   Masking is you pass a condition and then only elements satisfying the condition returns
*   It creates a copy.



```python
masking_matrix = np.round(10 * np.random.rand(5, 4))

masked_array = masking_matrix[masking_matrix > 5.]
print(masking_matrix)
print("------------------------")
print(masked_array)

masked_array = masking_matrix[(masking_matrix > 4.) & (masking_matrix < 8.)]
print("------------------------")
print(masked_array)
```

# np.vstack() — Vertical Stack (Row-wise)


*   Stacks arrays on top of each other
*   Arrays must have the same number of columns
*   1D arrays are treated as rows



```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.vstack((a, b))
print(result)
```

# np.hstack() — Horizontal Stack (Column-wise)


*   Stacks arrays side by side
*   Arrays must have the same number of rows
*   For 1D arrays, it concatenates along the only axis


```python
a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])
print(np.vstack((a, b)))
print("------------------------")
print(np.hstack((a, b)))
```
